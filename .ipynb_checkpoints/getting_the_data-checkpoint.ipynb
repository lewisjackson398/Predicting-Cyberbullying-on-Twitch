```python
# Web scraping, pickle imports
import requests
import pickle

# Scrapes chat data from https://overrustlelogs.net/
def url_to_chat(url):
    page = requests.get(url).text
    print(url)
    return page

# URLs of chats in scope
urls = ['https://overrustlelogs.net/Drdisrespect%20chatlog/February%202020/2020-02-28.txt',
        'https://overrustlelogs.net/Loltyler1%20chatlog/February%202020/2020-02-29.txt',
        'https://overrustlelogs.net/Nickmercs%20chatlog/February%202020/2020-02-28.txt',
        'https://overrustlelogs.net/Xqcow%20chatlog/February%202020/2020-02-29.txt',
        'https://overrustlelogs.net/Summit1g%20chatlog/February%202020/2020-02-29.txt',
        'https://overrustlelogs.net/Pokimane%20chatlog/February%202020/2020-02-29.txt',
        'https://overrustlelogs.net/Kaceytron%20chatlog/February%202020/2020-02-28.txt',
        'https://overrustlelogs.net/Sweet_anita%20chatlog/February%202020/2020-02-29.txt',
        'https://overrustlelogs.net/Xchocobars%20chatlog/February%202020/2020-02-22.txt',
        'https://overrustlelogs.net/Amouranth%20chatlog/February%202020/2020-02-29.txt'
       ]

# Comedian names
streamers = ["Drdisrespect", "Loltyler1", "Nickmercs", "Xqcow", "Summit1g", "Pokimane", "Kaceytron", "Sweet_anita", "Xchocobars", "Amouranth"]
```

```python
# # Actually request chat (takes a few minutes to run)
chats = [url_to_chat(u) for u in urls]
```

```python
# # Pickle files for later use

# # Make a new directory to hold the text files
# !mkdir transcripts

for i, c in enumerate(streamers):
    with open("raw_data/" + c + ".txt", "wb") as file:
        pickle.dump(chats[i], file)
```

```python
# Load pickled files
data = {}
for i, c in enumerate(streamers):
    with open("raw_data/" + c + ".txt", "rb") as file:
        data[c] = pickle.load(file)
```
